### Data Info: 
The data are records related to 80 patients' teeth health after receiving orthodontic treatment. Patients are divided to two groups of 40, one underwent orthodontic treatment with two differnt fixed retainers . For each patient, 6 different factors of teeth health for six teeth (3 same tooth number in both sides of the mounth) of front and back have been examined. The examination has been repeated for three time with time interval of 3 months. As a result, for each patient, 72 columns of data for each examination or in total three examination 26 fields of data have been collected which add to two other field of gender and age summed up to total 218 fields of data for each patient.  Six different factors can be abbreviated with ['CI','DI','GI','GR','PI','PPD']
Considering 3 examination, 3 teeth and two side of each tooth undergone the examination total fields of data for each patients are summarized as

```python


# the function plot distribution of data based on gender, group and age
import matplotlib.ticker as mticker # to customize ticks of the plot
def distribu_age_gender(df):
    sns.set_style("darkgrid")
    fig= plt.figure(figsize=(20,30)) # size of the figure
    """to have two subplots each group we use FacetGrid 
    (This class maps a dataset onto multiple axes arrayed in a grid of rows and columns 
    that correspond to levels of variables in the dataset)"""
    g = sns.FacetGrid(df, col="Group",
          height=10, aspect=2)  # Have two columns of plots for each group with aspect ration 2 for width/height 
    kwargs=dict(edgecolor="k", linewidth=2)
    g.map_dataframe(sns.histplot, x="Age", hue ='Gender',multiple="stack",discrete=True, shrink=.8, palette=colors, 
                    bins = max(df.Age)-min(df.Age), **kwargs)
   # to customize each axis in the plot 
    for i, ax in enumerate(g.axes.flatten()):
    #     ax.tick_params(axis='both', labelsize= 30, which = 'both')
        ax.tick_params(axis ='both', which ='both',
                   labelsize = 30, colors ='k')
    #     ax.grid(True)
        ax.legend(['Female', 'Male'], fontsize=30)
        ax.set_xlabel('Age', fontsize=30, color ='k')
        ax.set_ylabel('Count', fontsize=30, color ='k')
        ax.set_title(f'Group {i+1}', fontsize=50, color ='k')
        ticks_loc = list(range(13, 40)) # for naming the ticks of the x axis from the smallest age to the highest age
        ax.xaxis.set_major_locator(mticker.FixedLocator(ticks_loc))
    g.fig.suptitle('Age Distribution', size = 100, y = 1.3)

distribu_age_gender(df_data)

```

```python
def melted_main_factors(dict_df, main_factors):
    dict_mfactor = {}
    keys_dict = dict_df.keys()
    for mkey in main_factors:
        dict_mfactor[mkey] = pd.concat([dict_df[key].melt(id_vars = ['id','Gender','Age','Group', 'Check_up', 'Factor'],value_vars = dict_df[key].columns[6:])
         for key in keys_dict if dict_df[key].Factor.any() == mkey])

    return dict_mfactor    
    
```
